   ### 주요 파라미터
   - 키를 기준으로 DataFrame의 로우를 합친다. SQL이나 다른 관계형 데이터베이스의 join 연산과 동일함.
   - 주요 파라미터        
      . left, right : merge할 DataFrame 객체이름
      
      . how = 'inner', #left, right, outer
      
      . on = None, #merge의 기준이 되는 컬럼
      
      . left_on = None, #left DataFrame의 기준 컬럼
      
      . right_on = None, #right DataFrame의 기준 컬럼

```python
df_left = DataFrame({'KEY': ['k0', 'k1', 'k2', 'k3'],
                     'A': ['a0', 'a1', 'a2', 'a3'],
                     'B': ['b0', 'b1', 'b2', 'b3']})
df_right = DataFrame({'KEY': ['k2', 'k3', 'k4', 'k5'],
                     'C': ['c2', 'c3', 'c4', 'c5'],
                     'D': ['d2', 'd3', 'd4', 'd5']})
```

```python
df_left
```

```python
df_right
```

```python
# inner join
# 4 * 4 = 16번의 연산이 이루어짐
# key값이 같은 놈이 나옴 => k2,k3
pd.merge(df_left,df_right,how='inner') # inner join이란것을 알려주기 위해서
```

```python
# left outer join
# left에 있는 데이터는 일단 다 출력해달라 -> 없는 놈은 NaN 처리하더라도
pd.merge(df_left,df_right,how='left')
```

```python
#right outer join
pd.merge(df_left,df_right,how='right')
```

```python
# fully outer join
pd.merge(df_left,df_right,how='outer') # 둘 중 한군데라도 있으면 출력
```

### 실습

```python
# 실습 아래 2개의 DataFrame을 K1과 K2 컬럼으로 outer merge
df_left2 = DataFrame({'K1': ['k0', 'k1', 'k2', 'k3'],
                     'A': ['a0', 'a1', 'a2', 'a3'],
                     'B': ['b0', 'b1', 'b2', 'b3']})
df_right2 = DataFrame({'K2': ['k2', 'k3', 'k4', 'k5'],
                     'C': ['c2', 'c3', 'c4', 'c5'],
                     'D': ['d2', 'd3', 'd4', 'd5']})

pd.merge(df_left2,df_right2,left_on='K1',right_on='K2',how='outer')
```

```python
# 실제 데이터 실습
#1. 사용자 데이터 읽어오기
# 데이터 파일: 'data/movielens/users.dat'
# 컬럼명 = ['사용자아이디','성별','연령','직업','지역']
users = pd.read_csv('data/movielens/users.dat',
            sep='::', # 원래 sep가 1바이트여야하는데 2바이트라 warning
            header=None,
            names=['사용자아이디','성별','연령','직업','지역'],
            engine='python')

users.head()
```

```python
# 2. 평점 데이터 읽어오기
# 데이터 파일 : data/movielens/ratings.dat
# 컬럼명들은 ['사용자아이디', '영화아이디', '평점', '타임스탬프']
ratings = pd.read_csv('data/movielens/ratings.dat',
            sep='::', # 원래 sep가 1바이트여야하는데 2바이트라 warning
            header=None,
            names=['사용자아이디', '영화아이디', '평점', '타임스탬프'],
            engine='python')

ratings.head()
```

```python
# 3. 사용자와 평점 데이터 합치기
# 데이터를 애초에 왜 합쳐두지 않을까? 데이터를 더 효율적으로 관리하기 위해서
# 어떤 정보를 변경할때 나눠뒀으면 훨씬 편함
pd.merge(users,ratings,
         on = '사용자아이디',
         how = 'inner')
```

## 4-2. pandas.concat(axis = 0)

```python
df1 = DataFrame([[1,'aa'],[2,'bb']],columns = list('AB'))
df2 = DataFrame([[3,'cc'],[4,'dd']],columns = list('AB'))
```

```python
# df1과 df2 concat 수행하기
# 축 방향은 기본으로 0
# dataframe 컬럼은 같고 데이터만 다를때 이어붙이려는 함수로 많이 쓰임
pd.concat([df1,df2]) # list 형태로 줘야함
```

## 4-3. concat(axis = 1)

```python
# 샘플 데이터 생성
df1 = DataFrame([[1,10],[2,20], [3,30]], columns = list('AB'), index = ['aa','bb', 'cc'])
df2 = DataFrame([[3,'xx'],[4,'yy']], columns = list('CD'), index = ['aa','bb'])
```

```python
# concat with axis = 1 수행
pd.concat([df1, df2], axis = 1, sort=True)
```

```python
pd.concat([df1,df2],axis=1,sort=True,join='outer')
```
