## 1. Matplotlib 소개
- 파이썬에서 자료를 차트(chart)나 플롯(plot)으로 시각화(visulaization)하는 패키지
- Matplotlib는 다음과 같은 정형화된 차트나 플롯 이외에도 저수준 api를 사용한 다양한 시각화 기능 제공
- 다양한 시각화 도구를 제공함
    - 라인 플롯(line plot)
    - 스캐터 플롯(scatter plot)
    - 컨투어 플롯(contour plot)
    - 서피스 플롯(surface plot)
    - 바 차트(bar chart)
    - 히스토그램(histogram)
    - 박스 플롯(box plot)

[Matplotlib 갤러리 웹사이트](http://matplotlib.org/gallery.html)

```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# 기본 그리기
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.plot.html
sr = Series([1,2,4,8,16],index=list('abcde'))
sr
```

```python
plt.plot(sr)
```

```python
# 1-2. bar graph 그리기
import random
sr2 = Series(random.sample(range(1,100),5),index=list('abcde'))
sr2
```

```python
plt.bar(sr2.index,sr2.values)
```

```python
# 1-3 위 2개의 그래프를 하나의 그래프로 그리기
fig, ax = plt.subplots() # 현재 figure와 axes를 반환
ax.plot(sr,color='red')
plt.bar(sr2.index,sr2.values,alpha=0.3) # 누적하면 됨
plt.show()
```

```python
# 1-4 두 개 그래프를 서로 다른 Y축으로 적용
fig, ax1 = plt.subplots()
ax1.tick_params(axis='y',labelcolor='r')
ax2 = ax1.twinx() # x축을 공유함
ax2.bar(sr2.index,sr2.values,alpha=0.3)
ax1.plot(sr,color='r')
plt.show()
```

```python
# 2. 여러 개 그래프 그리기
## 하나의 figure에 여러개의 axes를 그리기
fig,ax_list = plt.subplots(2,2)

ax_list[0,0].plot(sr)
ax_list[0,0].set_title('Graph #1')
ax_list[0,0].set_xlabel('x1')
ax_list[0,0].set_ylabel('y1')

ax_list[0,1].hist(random.sample(range(1,100),50),bins=10,color='b',alpha=0.3)
ax_list[0,1].set_title('Graph #2')
ax_list[0,1].set_xlabel('x2')
ax_list[0,1].set_ylabel('y2')

ax_list[1,0].scatter(random.sample(range(1,100),50),random.sample(range(1,100),50),color='r')
ax_list[1,0].set_title('Graph #3')
ax_list[1,0].set_xlabel('x3')
ax_list[1,0].set_ylabel('y3')

ax_list[1,1].bar(sr2.index,sr.values,color='g')
ax_list[1,1].set_title('Graph #4')
ax_list[1,1].set_xlabel('x4')
ax_list[1,1].set_ylabel('y4')

plt.subplots_adjust(hspace=1,wspace=1)
plt.show()
```

### 3) 대규모 전쟁의  공격진영과 수비 진영의 숫자 시각화
#### 대규모 전쟁 = 공격과 수비 모두 합쳐서 10,000명 이상 참가한 전쟁

```python
big_battles = battles[battles.attacker_size + battles.defender_size >= 100]
big_battles.head()
```

```python
big_battles = big_battles[['name','attacker_size','defender_size']]
big_battles
```

```python
big_battles.set_index('name',inplace=True)
```

```python
big_battles.plot(kind='barh',
                 stacked=True)
```

### 4) (실습) 가문 별로 전쟁을 벌인 횟수 시각화

```python
# attacker_1 ~ defender_4 컬럼 활용
# 영화 평점 분석 실습 중 장르 통계 코드 참고

# 첫번째 방법
battles[['attacker_1', 'attacker_2', 'attacker_3', 'attacker_4',
         'defender_1','defender_2', 'defender_3', 'defender_4']].head()
```

```python
# 두번째 방법
battles.loc[:,'attacker_1':'defender_4'].head()
```

```python
# 세번째 방법
battles[battles.columns[5:13]].head()
```

```python
# 네번째 방법
cols = battles.columns[battles.columns.str.contains('attacker_[1-4]|defender_[1-4]')]
ex1 = battles[cols]
ex1.head()
```

```python
house_to_battle_count = Series()

for col in ex1.columns:
    battle_cnt = ex1[col].value_counts()
    house_to_battle_count = house_to_battle_count.add(battle_cnt,fill_value= 0)
    
house_to_battle_count
```

```python
house_to_battle_count.sort_values().plot(kind='barh')
plt.show()
```

### 5) (실습) 년도별로 사망자의 숫자와 전쟁이 벌어진 횟수 시각화

```python
# 연도별 사망자 수 구하기
death_count = death['Death Year'].value_counts().sort_index()
death_count
```

```python
# pivot_table 활용
death.pivot_table(index='Death Year',aggfunc='count',values='Name')
```

```python
# 연도별 전쟁 수
battle_count = battles['year'].value_counts().sort_index()
battle_count
```

```python
# pivot_table 활용
battles.pivot_table(index='year',aggfunc='count',values='name')
```

```python
death_battle_count = pd.concat([death_count,battle_count],axis=1)
death_battle_count.columns = ['# of death','# of battles']
death_battle_count.fillna(0,inplace=True)
death_battle_count.plot()
plt.show()
```
