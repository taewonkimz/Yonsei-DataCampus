# scikit-learn을 이용하여 붓꽃(iris) 품종 예측

```python
import sklearn
# 위와 같이 import하면 submodule 들은 함께 임포트 되지 않는다.
# 만약 모든 submodule을 다 임포트하고 싶으면, from sklearn import * 수행
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

# 1. iris 데이터 읽어오기
iris = load_iris()

# iris 데이터셋 확인
display(iris.keys())
display(iris.data)
```

```python
# iris.target = label(정답)
display(iris.target)
```

```python
# label(target)의 이름
iris.target_names
```

```python
iris.feature_names
```

```python
# iris_data = 피처(feature) 정보
iris_data = iris.data
```

```python
iris_label = iris.target
```

```python
# 붓꽃 데이터 세트를 DataFrame으로 변환
import pandas as pd
iris_df = pd.DataFrame(iris_data, columns = iris.feature_names)
```

```python
iris_df['label'] = iris.target
iris_df
```

```python
# 1) 결정트리 분류 객체 생성
dt_clf = DecisionTreeClassifier()

# 2) 학습
dt_clf.fit(iris.data, iris.target)

# 3) 평가
dt_clf.score(iris.data, iris.target)
```

```python
# 학습/검증용 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.3, random_state = 100)
```

```python
pd.Series(y_train).value_counts()
```

```python
# 2) 학습
dt_clf.fit(X_train, y_train)
```

```python
# 3) 평가
print('학습데이터 정확도 :', dt_clf.score(X_train, y_train))
print('테스트데이터 정확도 :', dt_clf.score(X_test, y_test))
```

```python
dt_pred = dt_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, dt_pred)
```

# 교차검증(Cross-Validation)

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits = 5, shuffle = False)
```

```python
accuracies = []

# kfold.split : kfold 수행
for train_idx, test_idx in kfold.split(iris.data, iris.target) :
    X_train, X_test = iris.data[train_idx], iris.data[test_idx]
    y_train, y_test = iris.target[train_idx], iris.target[test_idx]
    print(pd.Series(y_train).value_counts())
    dt_clf.fit(X_train, y_train)
    print(dt_clf.score(X_test, y_test))
```

```python
# Stratiied Kfold 수행
from sklearn.model_selection import StratifiedKFold
skfold = StratifiedKFold(n_splits = 5, shuffle = False)

for train_idx, test_idx in skfold.split(iris.data, iris.target) :
    X_train, X_test = iris.data[train_idx], iris.data[test_idx]
    y_train, y_test = iris.target[train_idx], iris.target[test_idx]
    print(pd.Series(y_train).value_counts())
    dt_clf.fit(X_train, y_train)
    print(dt_clf.score(X_test, y_test))
```
