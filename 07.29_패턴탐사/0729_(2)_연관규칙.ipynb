## 임의의 가상 트랜잭션 데이터 10개에 대하여 연관 규칙 탐사를 수행하시오
### - 최소 지지도 0.2, 최소 신뢰도 0.1로 지정하십시오
### - 데이터는 가상의 랜덤한 1~10이 있을 수 있습니다.
### - 각 트랜잭션의 길이는 최소 2, 최대 4라고 가정합니다.

```python
import random
```

```python
# random.sample(range(범위), 원하는 개수) : 범위 내에서 원하는 개수만큼 무작위로 선택(중복 X)
# random.choice([]) : 리스트 내의 숫자 중 무작위로 하나 선택

tr = []

for _ in range(10) :  # Dont` care
    tr_sub = random.sample(range(1,11), random.choice([2,3,4]))
    tr.append(tr_sub)
```

```python
tr
```

```python
import apyori as ap
```

```python
rules = ap.apriori(tr, min_support = 0.2, min_confidence = 0.1)
```

```python
# 연관규칙을 탐사한 트랜잭션은 탐사된 뒤 사라진다.
# 따라서 다시 탐사하고 싶다면 트랜잭션을 다시 만들어주어야 한다.
import pandas as pd
table = []

for rule in rules :
    support = rule[1]
    infor = list(rule[2])
    if len(infor) > 1 :
        for i in infor : 
            row = [set(i[0]), "==>", set(i[1]), support, i[2], i[3]]
            table.append(row)

# Left/Right Hand Side
colname = ["LHS", "==>", "RHS", "Support", "Confidenct", "Lift"]
result = pd.DataFrame(table, index = range(1, len(table)+1), columns = colname)
result
```

## market.csv 파일을 불러와서 연관 규칙 탐사 수행
### - 최소지지도를 0.01로, 최소신뢰도를 0.1로 지정하여 수행한다.
### - 이 결과를 바탕으로 다음의 질문에 대해 의사결정을 해본다.
### - soda를 구입하는 고객에게 root vegetables를 추천할 필요가 있는가?
### - beef를 구입하는 고객에게 root vegetables를 추천할 필요가 있는가?

```python
import pandas as pd
import apyori as ap
```

```python
f = open(r"C:\Users\김태원\Desktop\market.csv")

tr = list(ap.load_transactions(f, delimiter = ','))

f.close()
```

```python
rules = ap.apriori(tr, min_support = 0.01, min_confidence = 0.1)
```

```python
table = []

for rule in rules :
    support = rule[1]
    infor = list(rule[2])
    if len(infor) > 1 :
        for i in infor : 
            row = [set(i[0]), "==>", set(i[1]), support, i[2], i[3]]
            table.append(row)

# Left/Right Hand Side
colname = ["LHS", "", "RHS", "Support", "Confidence", "Lift"]
result = pd.DataFrame(table, index = range(1, len(table)+1), columns = colname)
result
```

```python
result1 = result.sort_values(by = "Lift", ascending = False)
final = result1[result['RHS'] == {"root vegetables"}]
display(final)  # beef는 추천, soda는 비추천
```
